### 选股池：主板标的

```python
import tushare as ts
pro = ts.pro_api()
stock_panel_l = pro.stock_basic(exchange = '',list_status = 'L',market = "主板")
stock_panel_D = pro.stock_basic(list_status = 'D',market = "主板")
stock_panel_P = pro.stock_basic(list_status = 'P',market = "主板")

def transfer_code(df):
    return df["ts_code"].apply(lambda x:x.split('.')[0]).values

all_valid = np.array(list(set(transfer_code(stock_panel_l)).\
union(set(transfer_code(stock_panel_D))).\
union(set(transfer_code(stock_panel_P)))))
```

##### 曾经在hs300或者zz500中出现过的成分股

```python
stock_pool_datahs300 = io1.load("HS_comp")
stock_pool_datazz500 = io1.load("ZZ_comp")
importlib.reload(Rf)
dynamic_stock_poolhs300 = Rf.component_select_code(stock_pool_datahs300,\
                                            start_time = pd.to_datetime("2018-01-01"),\
                                           end_time = pd.to_datetime("2021-12-10"))
dynamic_stock_poolzz500 = Rf.component_select_code(stock_pool_datazz500,\
                                            start_time = pd.to_datetime("2018-01-01"),\
                                           end_time = pd.to_datetime("2021-12-10"))
valid_codehs300 = Rf.extract_valid_code(dynamic_stock_poolhs300) #非科创板
valid_codezz500 = Rf.extract_valid_code(dynamic_stock_poolzz500)
valid_zbhs300 = Rf.get_valid_zb_code(all_valid,valid_codehs300) #主板标的
valid_zbzz500 = Rf.get_valid_zb_code(all_valid,valid_codezz500)
```

### 行业信息
##### 使用新的申万行业分类31个一级行业

```python
df = pro.index_classify(level='L1', src='SW2021')

all_constituent = pd.DataFrame([])
for row,value in df.iterrows():
    index_name = value["index_code"]
    temp = pro.index_member(index_code = index_name)
    all_constituent = pd.concat([all_constituent,temp],axis = 0)

#有效选股池，有些是曾经的成分股已经退市
all_constituent["code"] = all_constituent["con_code"].apply(lambda x:x.split('.')[0])
all_constituent1 = pd.merge(all_constituent,df[["index_code","industry_name"]],left_on = "index_code",\
                           right_on = "index_code")
hs300_valid = all_constituent1[pd.Index(all_constituent1["code"]).isin(valid_zbhs300)][["code","index_code","industry_name"]]
zz500_valid = all_constituent1[pd.Index(all_constituent1["code"]).isin(valid_zbzz500)][["code","index_code","industry_name"]]

#所有有效标的池（在valid_code_非创业同时在行业信息有数据的标的）
valid_code = io1.load("valid_code_chuangye")
res = all_constituent1[pd.Index(all_constituent1["code"]).isin(valid_code)][["code","index_code","industry_name"]]
```
